```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def convert(string):
    li=string.split(",")
    ans=""
    for i in range(len(li)):
        ans=ans+li[i]
    temp=int(ans)
    return temp
ans=pd.read_csv("startup_funding.csv")
data=ans.copy()
cities=data["CityLocation"]
dictt={}
data
data.fillna("unknown",inplace=True)
nextdata=[]
for ele in cities:
    if ele=="unknown":
        nextdata.append(ele)
        dictt[ele]=dictt.get(ele,0)+1
        continue
    if "bangalore" in ele:
        ele="Bangalore"
        nextdata.append(ele)
        dictt[ele]=dictt.get(ele,0)+1
        continue
    if "Delhi" in ele:
        ele="New Delhi"
        nextdata.append(ele)
        dictt[ele]=dictt.get(ele,0)+1
        continue
    if " /" in ele:
        li=ele.split()
        ele=li[0]
        nextdata.append(ele)
        dictt[ele]=dictt.get(ele,0)+1
        continue
    if "/" in ele:
        li=ele.split("/")
        ele=li[0]
        nextdata.append(ele)
        dictt[ele]=dictt.get(ele,0)+1
        continue
    nextdata.append(ele)
    dictt[ele]=dictt.get(ele,0)+1
city=["Bangalore","Mumbai","New Delhi","Gurgaon","Pune","Noida","Hyderabad","Chennai","Ahmedabad","Jaipur"]
data2=ans.copy()
amount=data2["AmountInUSD"]
money=[]
for i in range(len(amount)):
    ele=amount[i]
    ele=str(ele)
    if ele[0:len(ele)]=="nan":
        money.append(0)
        continue
    else:
        num=convert(ele)
        money.append(num)
ansdictt={}
for i in range(len(money)):
    ele=nextdata[i]
    add=money[i]
    if nextdata[i] in city:
        ele=nextdata[i]
        add=money[i]
        ansdictt[ele]=ansdictt.get(ele,0)+add
    
citypie=[]
sumpie=[]
for ele in ansdictt:
    citypie.append(ele)
    sumpie.append(ansdictt[ele])
explode=[0.1,0.2,0.3,0.1,0.5,0.4,0.6,0.7,0.7,0.5]
plt.pie(sumpie,labels=citypie,autopct="%.2f%%",explode=explode)
plt.axis("equal")
print("Bangalore","49.71")
print("New Delhi","16.63")
print("Mumbai","13.90")
print("Gurgaon","12.21")
print("Chennai","2.43")
print("Pune","2.16")
print("Hydreabad","1.15")
print("Noida","1.01")
print("Ahmedabad","0.58")
print("Jaipur","0.21")
```
