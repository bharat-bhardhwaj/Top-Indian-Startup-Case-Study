```python
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv("startup_funding.csv")
data=dataset.copy()
li=data["Date"]
dictt={}
for i in li:
    ele=i[-1:-5:-1]
    ele=ele[-1:-5:-1]
    dictt[ele]=dictt.get(ele,0)+1
year=[]
startups=[]
for i in dictt:
    year.append(i)
    startups.append(dictt[i])
year=year[len(year)-1::-1]
startups=startups[len(startups)-1::-1]
plt.plot(year,startups)
for i in range(len(year)):
    print(year[i],startups[i])
```
